```python
from PIL import Image
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
```

## Illustration 1: Climate change in Germany

```python
fig,ax=plt.subplots(1,2,figsize=(30,30))
rotate=0
im=Image.open("imp_1_2.png")
im=im.rotate(rotate, Image.NEAREST, expand=1)
ax[0].imshow(im)
im2=Image.open("orig_1.png")
im2=im2.rotate(rotate, Image.NEAREST, expand=1)
ax[1].imshow(im2)
plt.show()
ax[0].get_xaxis().set_visible(False)
ax[0].get_yaxis().set_visible(False)
ax[1].get_xaxis().set_visible(False)
ax[1].get_yaxis().set_visible(False)
```

## Comments on improvement of Graph 1

#### 1. The original graph(on the right) incorrectly joined the adta points to the original (0,0) 
#### misleading the user in terms of graph that Temp,Precipitation = (0,0). This was fixed in the original graph as we moved to bar plot

#### 2. Reduced cLutter of Graph Labels and ordered them on the left

#### 3. Remove x axis and y-axis

#### 4. Used color to drive home the conclusion that weatehr has been hotter. The colors on the right are distracting(without legend)

#### 5. Removed Gray background

#### 6. Used soothing colors to denote dry,wet,cold and hot so that the message sticks with the users, removing cognitive load

#### 7. This graph was tricky as we SHOULD NOT ideally use two axes. But we had 4 dimensional data points here, hence it was taken care of using colors in a way that user understands Cold& Dry, Cold & Wet and so on.
